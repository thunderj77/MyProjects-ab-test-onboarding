### 1. Загрузка исторических данных и их предобработка

На этом этапе необходимо загрузить исторические данные о действиях пользователей и подготовить их к анализу. Предобработка включает проверку структуры датасета, типов данных, обработку пропусков и дубликатов, а также приведение временных полей к стандартному формату.

````python
# Работа с данными и численные вычисления
import numpy as np
import pandas as pd

# Визуализация
import matplotlib.pyplot as plt
import seaborn as sns
from plotly import graph_objects as go

# Статистический анализ
from statsmodels.stats.proportion import proportions_ztest
from scipy import stats

pd.set_option('future.no_silent_downcasting', True) # убираем предупреждения о будущих изменениях
````

````python
def load_data(filename):
    # Загружаем данные без привязки к директории датасета
    # Варианты путей в порядке приоритета
    paths = [
        f'/datasets/{filename}',           # Путь на Практикуме
        f'datasets/{filename}',             # Папка datasets в текущей директории
        f'./datasets/{filename}',           # Явное указание текущей директории
        filename,                            # Файл в текущей директории
        f'https://code.s3.yandex.net/datasets/{filename}', # Путь на Облаке
    ]
    
    # Перебираем все возможные пути к файлу
    for path in paths:
        try:
            # Пытаемся прочитать CSV-файл по текущему пути
            df = pd.read_csv(path)
            # Если чтение успешно, выводим сообщение об успехе
            print(f"✅ Файл загружен: {path}")
            # Возвращаем загруженный датафрейм
            return df
        except FileNotFoundError:
            # Если файл не найден по этому пути, переходим к следующему
            continue
            
    # Если ни один путь не сработал, выбрасываем исключение с информативным сообщением
    raise FileNotFoundError(f"Не удалось найти файл {filename} ни по одному из путей")

# Загружаем исторические данные
hist = load_data('pa_sollmafin_hist.csv')
# Загружаем результаты А/B тестирования
abt = load_data('pa_sollmafin_abt.csv')
````

````python
# Обходим все таблицы: выводим структуру данных (info) и первые 3 строки для проверки
dataframes = {
    "hist": hist,
    "abt": abt
}

for name, df in dataframes.items():
    print(f"\n=== {name} ===")
    df.info()
    display(df.head(3))
````

### 3. Анализ воронок событий

**Что делаем:**
- Строим классическую воронку событий (шаги, обязательные для всех пользователей)
- Строим пошаговую воронку (шаги, которые пользователи могут пропускать)
- Выявляем узкие места и точки оттока пользователей

**Ожидаемый результат:** понимание пользовательского пути, идентификация этапов с наибольшей потерей аудитории для дальнейшей оптимизации.

````python
# Определяем шаги воронки
funnel_steps = [
    ('install / open_web', 'Установка'),
    ('introduction', 'Вступление'),
    ('registration', 'Регистрация'),
    ('onboarding_complete', 'Онбординг'),
    ('first_deposit', '1-й депозит'),
    ('asset_purchase', 'Покупка актива'),
    ('second_deposit', '2-й депозит')
]

# Считаем уникальных пользователей
users_count = hist.groupby('event_name')['user_id'].nunique()

# Формируем датафрейм
funnel_data = pd.DataFrame({
    'step_en': [step[0] for step in funnel_steps],
    'step_short': [step[1] for step in funnel_steps],
    'step_num': [f'{i+1}. {step[0]}' for i, step in enumerate(funnel_steps)],
    'users': [users_count.get(step[0], 0) for step in funnel_steps]
})

# Добавляем проценты
funnel_data['pct_from_start'] = (funnel_data['users'] / funnel_data['users'].iloc[0] * 100).round(1)
funnel_data['conv_from_prev'] = (funnel_data['users'] / funnel_data['users'].shift(1) * 100).round(1)

# classic_funnel для обратной совместимости (если нужно)
classic_funnel = funnel_data[['step_num', 'users', 'pct_from_start']].copy()
classic_funnel.columns = ['step_name', 'users', 'pct_from_start']
````

````python
# Создаем интерактивную воронку
fig = go.Figure()

fig.add_trace(go.Funnel(
    name='Воронка конверсии',
    y=funnel_data['step_short'].tolist(),
    x=funnel_data['users'].tolist(),
    textposition="inside",
    textinfo="value+percent initial",
    texttemplate='<b>%{value:,.0f} чел.</b><br><i>%{percentInitial:.1%}</i>',
    marker=dict(
        color=['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728', '#9467bd', '#8c564b', '#e377c2'],
        line=dict(width=2, color='white')
    ),
    connector=dict(line=dict(color='gray', width=1, dash='dot'))
))

# Добавляем аннотации с конверсией от предыдущего этапа
for i in range(len(funnel_data)-1):
    conv_rate = funnel_data.iloc[i+1]['conv_from_prev']
    
    # Определяем цвет в зависимости от величины падения
    if conv_rate < 50:
        color = 'red'
    elif conv_rate < 80:
        color = 'orange'
    else:
        color = 'green'
    
    fig.add_annotation(
        x=0.85,
        y=len(funnel_data) - i - 1.5,
        text=f'<b>{conv_rate:.1f}%</b>',
        showarrow=True,
        arrowhead=2,
        arrowcolor=color,
        arrowsize=1.5,
        arrowwidth=2,
        font=dict(size=13, color=color, family='Arial Black'),
        bgcolor='white',
        bordercolor='lightgray',
        borderwidth=1,
        borderpad=4
    )

# Добавляем критическое падение текстом
critical_step = 'Покупка актива'  # Самый большой провал
critical_rate = funnel_data[funnel_data['step_short'] == critical_step]['conv_from_prev'].values[0]

fig.update_layout(
    title={
        'text': f'<b>Воронка конверсии пользователей</b><br><sup>⏬ Проценты справа — конверсия от предыдущего этапа</sup>',
        'x': 0.5,
        'xanchor': 'center',
        'font': {'size': 18, 'family': 'Arial'}
    },
    height=650,
    showlegend=False,
    annotations=[
        dict(
            x=0.5,
            y=-0.1,
            xref='paper',
            yref='paper',
            text=f'<b>⚠️ Ключевая точка оттока:</b> {critical_step} → падение на {100-critical_rate:.1f}% пользователей',
            showarrow=False,
            font=dict(size=12, color='red'),
            bgcolor='#ffe6e6',
            borderpad=10,
            bordercolor='red',
            borderwidth=1
        )
    ]
)

fig.show()
````

````python
# Создаем funnel_df из classic_funnel
funnel_df = pd.DataFrame({
    'step': classic_funnel['step_name'].str.split('. ').str[1],
    'users': classic_funnel['users'],
    '% от предыдущего': (classic_funnel['users'] / classic_funnel['users'].shift(1) * 100).round(1)
})

# Динамика потерь и конверсий
stages = funnel_df['step'].tolist()
funnel_users = funnel_df['users'].tolist()
drops = [funnel_users[i-1] - funnel_users[i] for i in range(1, len(funnel_users))]

# Каскадная диаграмма потерь и конверсия между этапами
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Левая часть: каскад потерь
x_pos = np.arange(len(stages))

# Цвета для потерь (чем больше потеря, тем краснее)
drop_colors = []
for i in range(len(drops)):
    pct = (drops[i] / funnel_users[i]) * 100
    if pct < 10:
        drop_colors.append('#2ecc71')      # зеленый - малые потери
    elif pct < 20:
        drop_colors.append('#f39c12')      # оранжевый - средние
    else:
        drop_colors.append('#e74c3c')      # красный - критические

# Столбцы с пользователями
ax1.bar(x_pos, funnel_users, color='#3498db', alpha=0.8, zorder=2, label='Пользователи')

# Столбцы с потерями
for i in range(len(drops)):
    ax1.bar(x_pos[i+1], drops[i], bottom=funnel_users[i+1], color=drop_colors[i], 
            alpha=0.6, zorder=2, label='Потери' if i == 0 else '')

# Подписи на барах
for i, (stage, user) in enumerate(zip(stages, funnel_users)):
    ax1.text(i, user/2, f'{user:,}', ha='center', va='center', 
            fontweight='bold', color='white', fontsize=10)

# Подписи потерь
for i, drop in enumerate(drops):
    if drop > 0:
        pct = (drop / funnel_users[i]) * 100
        ax1.text(i+1, funnel_users[i+1] + drop/2, f'-{pct:.1f}%\n({drop:,})', 
                ha='center', va='center', fontweight='bold', color='darkred', fontsize=8)

ax1.set_xticks(x_pos)
ax1.set_xticklabels(stages, rotation=45, ha='right')
ax1.set_ylabel('Количество пользователей')
ax1.set_title('Каскадная диаграмма потерь', fontweight='bold', fontsize=13)
ax1.grid(axis='y', alpha=0.3)
ax1.set_ylim(0, max(funnel_users) * 1.3)
ax1.legend(loc='upper right')

# Правая часть: конверсия между этапами
conv_rates = funnel_df['% от предыдущего'].tolist()[1:]  # без первого (100%)
conv_stages = [f'{stages[i]} → {stages[i+1]}' for i in range(len(stages)-1)]

# Цветовая кодировка конверсий
colors_conv = []
for x in conv_rates:
    if x < 50:
        colors_conv.append('#e74c3c')      # красный - критически низкая
    elif x < 80:
        colors_conv.append('#f39c12')      # оранжевый - низкая
    else:
        colors_conv.append('#2ecc71')      # зеленый - хорошая

bars = ax2.barh(conv_stages, conv_rates, color=colors_conv, alpha=0.8)

# Подписи процентов на барах
for bar, rate in zip(bars, conv_rates):
    ax2.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, 
            f'{rate:.1f}%', va='center', fontweight='bold', fontsize=11)

# Линии порогов
ax2.axvline(x=50, color='red', linestyle='--', alpha=0.5, linewidth=1, 
            label='Критический порог (50%)')
ax2.axvline(x=80, color='orange', linestyle='--', alpha=0.5, linewidth=1, 
            label='Целевой порог (80%)')

ax2.set_xlabel('Конверсия (%)')
ax2.set_title('Конверсия между этапами', fontweight='bold', fontsize=13)
ax2.set_xlim(0, 105)
ax2.legend(loc='lower right')

plt.tight_layout()
plt.show()
````

### 4. Влияние уровня риска актива на открытие второго депозита

**Цель:** изучить, как уровень риска первого купленного актива влияет на вероятность открытия второго депозита.

**Гипотезы:**
- Пользователи, начинающие с low‑risk активов, чувствуют себя увереннее и чаще возвращаются
- High‑risk активы могут отпугивать начинающих инвесторов, снижая retention

**Метод:** группировка пользователей по уровню риска первого актива (low/medium/high) с расчётом конверсии во второй депозит для каждой группы.


````python
# Фильтруем события: только покупки активов и вторые депозиты
events_of_interest = ['asset_purchase', 'second_deposit']
hist_rel = hist[hist['event_name'].isin(events_of_interest)].copy()

# Отделяем покупки и вторые депозиты
purchases = hist_rel[hist_rel['event_name'] == 'asset_purchase'].copy()
deposits2 = hist_rel[hist_rel['event_name'] == 'second_deposit'][['user_id']].drop_duplicates()
deposits2['has_second_deposit'] = True

# Берём первый купленный актив каждого пользователя
first_purchase = (
    purchases
    .sort_values('event_ts')
    .groupby('user_id')
    .first()
    .reset_index()[['user_id', 'risk_level', 'asset', 'amount']]
)

# Добавляем флаг второго депозита (False, если не было)
user_risk_to_deposit = first_purchase.merge(
    deposits2, on='user_id', how='left'
).fillna({'has_second_deposit': False})

# Проверяем данные
print("Пример данных:")
display(user_risk_to_deposit.head())

print("\nСтатистика по risk_level (до заполнения NaN):")
print(first_purchase['risk_level'].value_counts(dropna=False))

print("\nКоличество пользователей без risk_level:")
print(first_purchase['risk_level'].isna().sum())
````

````python
# Группируем по уровню риска: считаем пользователей и вторые депозиты
funnel_by_risk = (
    user_risk_to_deposit
    .groupby('risk_level')
    .agg(
        total_users=('user_id', 'count'),
        second_deposit=('has_second_deposit', 'sum')
    )
    .assign(
        conversion_to_second_deposit_pct=lambda x: 
            (x['second_deposit'] / x['total_users'] * 100)
    )
    .reset_index()
)

# Принудительно преобразуем в float и округляем
funnel_by_risk['conversion_to_second_deposit_pct'] = (
    funnel_by_risk['conversion_to_second_deposit_pct'].astype(float).round(1)
)

display(funnel_by_risk)
````

**Наблюдаемая закономерность:** чем ниже риск первого актива, тем выше конверсия во второй депозит.

**Ключевые факты:**
- Пользователи с low‑risk активом возвращаются в **1.76 раза чаще**, чем с high‑risk
- Medium‑risk показывает результат, близкий к low‑risk – умеренный риск не отпугивает
- **High‑risk на старте – главный фактор снижения retention**

**Интерпретация:**
Пользователи, выбирающие менее рискованные активы, чувствуют себя увереннее и с большей вероятностью продолжают инвестировать. Альтернативное объяснение: высокорисковые активы (опционы, криптовалюта) требуют более осознанного подхода и реже провоцируют импульсивные повторные покупки.

**Рекомендация:** для удержания пользователей стоит либо мягко направлять новичков в low/medium‑рисковые инструменты, либо усиливать образовательный контент перед покупкой high‑risk активов.

### 2. Сравнение воронок событий

**Цель:** сравнить пользовательский путь в контрольной и тестовой группах, выявить влияние нового онбординга на конверсию между этапами.

**Метод:** построение классической и пошаговой воронки с визуализацией.

**Ожидаемый результат:** понимание, на каких этапах новый онбординг улучшает или ухудшает конверсию.

````python
# Этапы в нужном порядке
funnel_steps = [
    'install / open_web',
    'introduction',
    'registration',
    'onboarding_complete',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

# Фильтруем и сортируем события
abt_rel = abt[abt['event_name'].isin(funnel_steps)].copy()

# Для каждого пользователя — его путь
user_paths = (
    abt_rel
    .sort_values(['user_id', 'event_ts'])
    .groupby('user_id')
    .agg(
        group=('group', 'first'),
        path=('event_name', lambda x: list(x.drop_duplicates()))
    )
    .reset_index()
)

# Функция: до какого этапа дошёл пользователь
def get_max_step(path):
    for i, step in enumerate(reversed(funnel_steps)):
        if step in path:
            return len(funnel_steps) - i
    return 0

user_paths['max_step'] = user_paths['path'].apply(get_max_step)

# Агрегация по группам
funnel_by_group = {}
for group in ['control', 'test']:
    steps = user_paths[user_paths['group'] == group]['max_step']
    counts = steps.value_counts().reindex(range(1, 8), fill_value=0).sort_index()
    
    # Классическая воронка: >= N
    classic = {}
    cumsum = 0
    for step in range(7, 0, -1):
        cumsum += counts.get(step, 0)
        classic[step] = cumsum
    funnel_by_group[group] = [classic[i] for i in range(1, 8)]

# Собираем в DataFrame
funnel_comp = pd.DataFrame({
    'step_name': [
        '1. install / open_web',
        '2. introduction',
        '3. registration',
        '4. onboarding_complete',
        '5. first_deposit',
        '6. asset_purchase',
        '7. second_deposit'
    ],
    'control': funnel_by_group['control'],
    'test': funnel_by_group['test']
})

# Конверсии (пошаговые)
for group in ['control', 'test']:
    funnel_comp[f'{group}_conv_to_next_%'] = (
        funnel_comp[group].shift(-1) / funnel_comp[group] * 100
    ).round(1)

display(funnel_comp)
````

### 3. Влияние уровня риска актива на конверсию во второй депозит

**Гипотеза эксперимента:** новый онбординг может по-разному влиять на пользователей в зависимости от того, какой уровень риска они выбрали для первой инвестиции. Пользователи, начинающие с low‑risk активов, могут чувствовать себя увереннее, а high‑risk – наоборот, испытывать «постинвестиционный стресс».

**Цель анализа:** оценить, как новый онбординг изменил конверсию во второй депозит в разрезе уровней риска (low / medium / high), и определить, является ли это изменение статистически значимым.

**Метод:** для каждого уровня риска сравнивается доля пользователей, совершивших второй депозит, между контрольной и тестовой группами. Используется z-тест для пропорций.

**Гипотезы (для каждого уровня риска):**

- **H₀ (нулевая):** конверсия во второй депозит в тестовой группе не отличается от контрольной  
- **H₁ (альтернативная):** конверсия во второй депозит в тестовой группе отличается от контрольной  

**Уровень значимости:** α = 0.05

````python
# Только покупки активов
purchases = abt[abt['event_name'] == 'asset_purchase'][['user_id', 'group', 'risk_level', 'event_ts']].copy()

# Первый актив по времени
first_purchase = (
    purchases
    .sort_values('event_ts')
    .groupby('user_id')
    .first()
    .reset_index()[['user_id', 'group', 'risk_level']]
)

# Факт второго депозита
second_dep_users = abt[abt['event_name'] == 'second_deposit']['user_id'].unique()
first_purchase['has_second_deposit'] = first_purchase['user_id'].isin(second_dep_users)

# Агрегация по группам и risk_level
funnel_risk_ab = (
    first_purchase
    .groupby(['group', 'risk_level'])
    .agg(
        users=('user_id', 'count'),
        second_deposit=('has_second_deposit', 'sum')
    )
    .reset_index()
)

funnel_risk_ab['conversion_%'] = (
    funnel_risk_ab['second_deposit'] / funnel_risk_ab['users'] * 100
).round(1)

display(funnel_risk_ab)
````

Подготовлены данные о первом купленном активе каждого пользователя и факте второго депозита, сгруппированные по группе и уровню риска. Видна разница в конверсиях между control и test – особенно для high‑risk активов.

````python
# Фактический расчёт
print("Распределение risk_level по группам:")
display(first_purchase.groupby(['group', 'risk_level']).size().unstack(fill_value=0))

print("\nКонверсия во второй депозит по risk_level и группе:")
display(funnel_risk_ab.pivot(index='risk_level', columns='group', values='conversion_%'))
````

Распределение пользователей по уровням риска между группами различается – в тесте больше low‑risk (496 vs 285) и medium‑risk (744 vs 595), меньше high‑risk (737 vs 893). Конверсия во второй депозит в тестовой группе выше по всем трём категориям, особенно заметно для high‑risk (+18.3 п.п.).

На графике видим что новый онбординг наиболее заметно повысил конверсию во второй депозит в сегменте high‑risk (столбцы визуально различаются сильнее всего), тогда как для medium‑risk эффект минимален.

````python
results = []
for risk in ['low', 'medium', 'high']:
    data = funnel_risk_ab[funnel_risk_ab['risk_level'] == risk]
    count = data['second_deposit'].values
    nobs = data['users'].values
    stat, pval = proportions_ztest(count, nobs)
    results.append({
        'risk_level': risk,
        'control_conv': data[data['group']=='control']['conversion_%'].iloc[0],
        'test_conv': data[data['group']=='test']['conversion_%'].iloc[0],
        'delta_pp': data[data['group']=='test']['conversion_%'].iloc[0] - 
                    data[data['group']=='control']['conversion_%'].iloc[0],
        'p_value': pval,
        'significant': pval < 0.05
    })

pd.DataFrame(results).round({'p_value': 4})
````

### 4. Анализ метрик A/B-эксперимента

**Цель:** оценить влияние нового онбординга на ключевые монетизационные метрики.

#### Анализируемые метрики

| Метрика | Тип | Описание |
|---------|-----|----------|
| **Средний депозит на пользователя** | Ключевая | ARPU (включая нулевые депозиты) |
| **Конверсия регистрация → первый депозит** | Барьерная | Доля зарегистрировавшихся, внесших первый депозит |
| **Конверсия первый → второй депозит** | Вспомогательная 1 | Доля платящих, совершивших повторный депозит |
| **Средний депозит на платящего** | Вспомогательная 2 | Средний чек платящего пользователя |

#### План анализа

1. Рассчитать значения метрик для control и test
2. Оценить абсолютное и относительное изменение
3. Проверить статистическую значимость различий
4. Для ключевой метрики – построить кумулятивный ARPU и динамику p-value


````python
# Уникальные пользователи и их группы
users = abt[['user_id', 'group']].drop_duplicates()

# Сумма всех депозитов по пользователю
deposits = abt[abt['event_name'].isin(['first_deposit', 'second_deposit'])]
user_deposit = (
    deposits
    .groupby('user_id')['amount']
    .sum()
    .reset_index(name='total_deposit')
)

# Факты событий
has_first = abt[abt['event_name'] == 'first_deposit'][['user_id']].assign(has_first=1)
has_second = abt[abt['event_name'] == 'second_deposit'][['user_id']].assign(has_second=1)
has_reg = abt[abt['event_name'] == 'registration'][['user_id']].assign(has_reg=1)

# Объединяем всё
user_metrics = (
    users
    .merge(user_deposit, on='user_id', how='left')
    .fillna({'total_deposit': 0})
    .merge(has_reg, on='user_id', how='left').fillna({'has_reg': 0})
    .merge(has_first, on='user_id', how='left').fillna({'has_first': 0})
    .merge(has_second, on='user_id', how='left').fillna({'has_second': 0})
)

# Только зарегистрировавшиеся (для барьерной метрики)
registered = user_metrics[user_metrics['has_reg'] == 1]

display(user_metrics.head())
````

Данные подготовлены, каждый пользователь описан набором бинарных флагов и суммой депозитов, что позволяет рассчитывать любые метрики групповыми агрегациями.

````python
# 1. Ключевая метрика: средний депозит на ВСЕХ пользователей
key_metric = user_metrics.groupby('group')['total_deposit'].agg(['mean', 'count']).round(2)

# 2. Барьерная: конверсия reg → first_deposit
barrier_metric = (
    registered
    .groupby('group')
    .agg(
        reg=('user_id', 'count'),
        first_dep=('has_first', 'sum')
    )
    .assign(conv_reg_to_first=lambda x: (x['first_dep'] / x['reg'] * 100).round(1))
)

# 3. Вспомогательная 1: first → second
first_deposit_users = user_metrics[user_metrics['has_first'] == 1]
aux1 = (
    first_deposit_users
    .groupby('group')
    .agg(
        first=('user_id', 'count'),
        second=('has_second', 'sum')
    )
    .assign(conv_first_to_second=lambda x: (x['second'] / x['first'] * 100).round(1))
)

# 4. Вспомогательная 2: средний депозит на платящих
paying_users = user_metrics[user_metrics['total_deposit'] > 0]
aux2 = paying_users.groupby('group')['total_deposit'].mean().round(2)

# Собираем всё
metrics = pd.DataFrame({
    'group': ['control', 'test'],
    'key_avg_deposit_all': key_metric['mean'].values,
    'barrier_conv_reg_to_first': barrier_metric['conv_reg_to_first'].values,
    'aux1_conv_first_to_second': aux1['conv_first_to_second'].values,
    'aux2_avg_deposit_paying': aux2.values
})

# Добавляем сравнение
metrics = metrics.set_index('group')
metrics.loc['abs'] = metrics.loc['test'] - metrics.loc['control']
metrics.loc['rel, %'] = (metrics.loc['test'] / metrics.loc['control'] - 1) * 100

# Округляем
metrics = metrics.round({
    'key_avg_deposit_all': 2,
    'barrier_conv_reg_to_first': 1,
    'aux1_conv_first_to_second': 1,
    'aux2_avg_deposit_paying': 2,
    'rel, %': 1
})

metrics
````

**Предварительное наблюдение:** тестовая группа показывает улучшение по двум вспомогательным метрикам (повторная конверсия и средний чек), однако ARPU вырос незначительно, а конверсия в первый депозит снизилась. Для подтверждения значимости требуется статистическая проверка.

#### Статистические гипотезы для метрик A/B-теста

**1. Ключевая метрика — средний депозит на пользователя (ARPU)**
- Тест: двухвыборочный t-тест Уэлча (неравные дисперсии)
- H₀: средний депозит в группах одинаков
- H₁: средний депозит в тестовой группе отличается

**2. Барьерная метрика — конверсия регистрация → первый депозит**
- Тест: z-тест для долей
- H₀: конверсия reg → first_deposit одинакова в обеих группах
- H₁: конверсия в тестовой группе отличается

**3. Вспомогательная метрика 1 — конверсия первый → второй депозит**
- Тест: z-тест для долей
- H₀: конверсия first → second не изменилась
- H₁: конверсия в тестовой группе отличается

**4. Вспомогательная метрика 2 — средний депозит на платящего пользователя**
- Тест: двухвыборочный t-тест Уэлча
- H₀: средний депозит у платящих одинаков в группах
- H₁: средний депозит у платящих в тестовой группе отличается

**Уровень значимости для всех тестов:** α = 0.05

````python
# 1. Ключевая метрика — t-тест (среднее, непрерывная)
control_vals = user_metrics[user_metrics['group'] == 'control']['total_deposit']
test_vals = user_metrics[user_metrics['group'] == 'test']['total_deposit']
t_stat, p_key = stats.ttest_ind(test_vals, control_vals, equal_var=False)

# 2. Барьерная и вспомогательные 1 — z-тест для долей
def ztest_prop(count1, nobs1, count2, nobs2):
    return proportions_ztest([count2, count1], [nobs2, nobs1])

# barrier: reg → first
c_reg = barrier_metric.loc['control', 'reg']
c_first = barrier_metric.loc['control', 'first_dep']
t_reg = barrier_metric.loc['test', 'reg']
t_first = barrier_metric.loc['test', 'first_dep']
_, p_barrier = ztest_prop(c_first, c_reg, t_first, t_reg)

# aux1: first → second
c_first_u = aux1.loc['control', 'first']
c_second = aux1.loc['control', 'second']
t_first_u = aux1.loc['test', 'first']
t_second = aux1.loc['test', 'second']
_, p_aux1 = ztest_prop(c_second, c_first_u, t_second, t_first_u)

# aux2: t-тест по платящим
c_paying = paying_users[paying_users['group'] == 'control']['total_deposit']
t_paying = paying_users[paying_users['group'] == 'test']['total_deposit']
_, p_aux2 = stats.ttest_ind(t_paying, c_paying, equal_var=False)

# Итоговая таблица
significance = pd.DataFrame({
    'metric': [
        'Ключевая: avg deposit (all)',
        'Барьерная: reg → first_deposit',
        'Вспомог. 1: first → second',
        'Вспомог. 2: avg deposit (paying)'
    ],
    'p-value': [p_key, p_barrier, p_aux1, p_aux2],
    'significant (α=0.05)': [p_key < 0.05, p_barrier < 0.05, p_aux1 < 0.05, p_aux2 < 0.05]
}).round({'p-value': 4})

display(significance)
````

**Вывод:** новый онбординг **статистически значимо** повышает конверсию в повторный депозит и средний чек платящих. Влияние на ARPU и конверсию в первый депозит **статистически не значимо**.

````python
# 1. Добавим день события (не регистрации!)
abt['event_day'] = abt['event_ts'].dt.date

# 2. Сумма депозитов по: день — пользователь — группа
# Только депозиты
deposits_daily = abt[abt['event_name'].isin(['first_deposit', 'second_deposit'])].copy()

user_daily_revenue = (
    deposits_daily
    .groupby(['event_day', 'user_id', 'group'])['amount']
    .sum()
    .reset_index(name='daily_revenue')
)

# 3. Все пользователи (включая тех, кто не внёс депозит в этот день)
# Чтобы ARPU = сумма / всех пользователей группы, зарегистрировавшихся ДО или В день события
all_users = abt[['user_id', 'group', 'first_dt']].drop_duplicates()

# Сортируем дни событий
days = sorted(user_daily_revenue['event_day'].unique())

results = []

for day in days:
    # Накопленные данные: события до day включительно
    # и пользователи, зарегистрировавшиеся до day включительно
    users_up_to_day = all_users[all_users['first_dt'] <= day]
    
    # Выручка по пользователям (для тех, у кого не было депозита — 0)
    revenue_up_to_day = (
        user_daily_revenue[user_daily_revenue['event_day'] <= day]
        .groupby(['user_id', 'group'])['daily_revenue']
        .sum()
        .reset_index()
    )
    
    # Присоединяем ко всем пользователям
    full = users_up_to_day.merge(
        revenue_up_to_day, 
        on=['user_id', 'group'], 
        how='left'
    ).fillna({'daily_revenue': 0})
    
    # ARPU по группам
    control_arpu = full[full['group'] == 'control']['daily_revenue'].mean()
    test_arpu = full[full['group'] == 'test']['daily_revenue'].mean()
    
    # t-тест (двухвыборочный, неравные дисперсии)
    c_vals = full[full['group'] == 'control']['daily_revenue']
    t_vals = full[full['group'] == 'test']['daily_revenue']
    _, p_val = stats.ttest_ind(t_vals, c_vals, equal_var=False)
    
    results.append({
        'day': day,
        'control_arpu': control_arpu,
        'test_arpu': test_arpu,
        'delta_arpu': test_arpu - control_arpu,
        'p_value': p_val
    })

cumulative = pd.DataFrame(results)
````

Ступенька после 16 июня связана с завершением набора новых пользователей — дальнейший рост ARPU обеспечивают повторные вложения уже активной аудитории.

**Вывод:** к окончанию эксперимента статистически значимых различий в ARPU между группами не зафиксировано (p-value = 0.338 > 0.05).

````python
np.random.seed(341)
n_iter = 5000

# Словари для хранения
boot_results = {}
boot_diffs = {}  # Сохраняем все бутстрап-разницы для графиков

for q in [25, 50, 75]:
    diffs = []
    for _ in range(n_iter):
        # Бутстрап выборки с возвратом
        boot_c = np.random.choice(control_deposits, size=len(control_deposits), replace=True)
        boot_t = np.random.choice(test_deposits, size=len(test_deposits), replace=True)
        
        q_c = np.percentile(boot_c, q)
        q_t = np.percentile(boot_t, q)
        
        diffs.append(q_t - q_c)  # test - control
    
    diffs = np.array(diffs)
    boot_diffs[q] = diffs  # Сохраняем для графиков
    
    ci_low, ci_high = np.percentile(diffs, [2.5, 97.5])
    obs_diff = np.percentile(test_deposits, q) - np.percentile(control_deposits, q)
    
    boot_results[q] = {
        'control_p': np.percentile(control_deposits, q),
        'test_p': np.percentile(test_deposits, q),
        'obs_diff': obs_diff,
        'ci_low': ci_low,
        'ci_high': ci_high,
        'significant': not (ci_low <= 0 <= ci_high)
    }

# Собираем в DataFrame
results_df = pd.DataFrame(boot_results).T
results_df.index.name = 'percentile'
results_df = results_df.round(2)
print("Результаты бутстрапа:")
display(results_df)
````
